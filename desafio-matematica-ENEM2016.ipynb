```python
import numpy as np
import pandas as pd
def leitura_treino_teste():
  
  #train = pd.read_csv("/content/drive/My Drive/Mestrado/Projetos/Github/desafio-matematica-ENEM2016/train.csv")
  train = pd.read_csv("https://raw.githubusercontent.com/eduardodut/desafio-matematica-ENEM2016/master/train.csv")
  train = train.set_index('NU_INSCRICAO')

  #test =  pd.read_csv("/content/drive/My Drive/Mestrado/Projetos/Github/desafio-matematica-ENEM2016/test.csv")
  test = pd.read_csv("https://raw.githubusercontent.com/eduardodut/desafio-matematica-ENEM2016/master/test.csv")
  test = test.set_index('NU_INSCRICAO')
 

  return train, test

```

```python
def dicionario_codigo_prova(codigo):
    append_str = codigo+'_'

    string = np.fromstring(np.arange(train[codigo].unique().shape[0]).tostring(),dtype=int)

    # Append suffix / prefix to strings in list 
    pre_res = [append_str.lower() + str(sub) for sub in string] 

    return {k: v for k, v in zip(train[codigo].unique(),pre_res)}

print()
```

```python
def preprocessamento(train,test):
 

  train_copy = train.copy()
  test_copy = test.copy()
  lista_descarte = []


  def replace_dummy_remove_column(coluna, dicionario, train_copy, test_copy):
    train_replaced = train_copy.replace({coluna: dicionario})
    test_replaced = test_copy.replace({coluna: dicionario})
    return dummy_remove_column(coluna, train_replaced, test_replaced)

  def dummy_remove_column(coluna, train_copy, test_copy):
    return pd.concat([train_copy,pd.get_dummies(train_copy[coluna])],axis=1), pd.concat([test_copy,pd.get_dummies(test_copy[coluna])],axis=1)
  
  '''  
  NU_INSCRICAO: Número de inscrição 
  > Identificador de inscrição do candidato. será o indexador do dataset

  CO_UF_RESIDENCIA: Código da Unidade da Federação de residência
  > Será descartado
  '''
    
  lista_descarte.append('CO_UF_RESIDENCIA')
 
  '''
  SG_UF_RESIDENCIA: Sigla da Unidade da Federação de residência
  '''
  train_copy = pd.concat([train_copy,pd.get_dummies(train_copy['SG_UF_RESIDENCIA'])],axis=1)
  test_copy = pd.concat([test_copy,pd.get_dummies(test_copy['SG_UF_RESIDENCIA'])],axis=1)
  lista_descarte.append('SG_UF_RESIDENCIA')

  '''
  NU_IDADE: idade do candidato

  '''

  '''
  TP_SEXO: sexo do candidato
  '''
  train_copy = pd.concat([train_copy,pd.get_dummies(train_copy['TP_SEXO'])],axis=1)
  test_copy = pd.concat([test_copy,pd.get_dummies(test_copy['TP_SEXO'])],axis=1)
  lista_descarte.append('TP_SEXO')


  '''
  TP_COR_RACA: cor/raça do candidato 
  '''
  dic_TP_COR_RACA =  {0:"cor_raca_n_declarada",
                      1:"cor_raca_branca",
                      2:"cor_raca_preta",
                      3:"cor_raca_parda",
                      4:"cor_raca_amarela",
                      5:"cor_raca_indigena"}
  
  train_copy, test_copy = replace_dummy_remove_column("TP_COR_RACA",dic_TP_COR_RACA, train_copy, test_copy)
  lista_descarte.append('TP_COR_RACA')

  '''
  TP_NACIONALIDADE: Nacionalidade
  > 
  '''
  dic_TP_NACIONALIDADE =  {0:"nacionalidade_n_informada",
                      1:"nacionalidade_brasileira",
                      2:"nacionalidade_brasileira_natu",
                      3:"nacionalidade_estrang",
                      4:"nacionalidade_brasileira_exterior"}


  train_copy, test_copy = replace_dummy_remove_column('TP_NACIONALIDADE',dic_TP_NACIONALIDADE, train_copy, test_copy)
  lista_descarte.append('TP_NACIONALIDADE')
  '''
  TP_ST_CONCLUSAO: Situação de conclusão do Ensino Médio
  > 
  '''

  dic_TP_ST_CONCLUSAO = {0:"conclusao_n_informada",
                         1:"conclusao_concluido",
                         2:"conclusao_cursando_em_2016",
                         3:"conclusao_cursando_apos_2016",
                         4:"conclusao_n_concluido_cursando"}
  
  train_copy, test_copy = replace_dummy_remove_column('TP_ST_CONCLUSAO',dic_TP_ST_CONCLUSAO, train_copy, test_copy)
  lista_descarte.append('TP_ST_CONCLUSAO')
  '''
  TP_ANO_CONCLUIU: Ano de Conclusão do Ensino Médio
  > 
  '''
  dic_TP_ANO_CONCLUIU = {0:"ano_conclusao_n_informado",
                         1:"ano_conclusao_2015",
                         2:"ano_conclusao_2014",
                         3:"ano_conclusao_2013",
                         4:"ano_conclusao_2012",
                         5:"ano_conclusao_2011",
                         6:"ano_conclusao_2010",
                         7:"ano_conclusao_2009",
                         8:"ano_conclusao_2008",
                         9:"ano_conclusao_2007",
                         10:"ano_conclusao_anterior_2007"}

  train_copy, test_copy = replace_dummy_remove_column('TP_ANO_CONCLUIU',dic_TP_ANO_CONCLUIU, train_copy, test_copy)
  lista_descarte.append('TP_ANO_CONCLUIU')
  '''
  TP_ESCOLA:Tipo de escola do Ensino Médio
  > 
  Há apenas uma observação do tipo 4, portanto será tratada como tipo 3
  '''
  dic_TP_ESCOLA = {1:"tp_escola_n_informada",
                   2:"tp_escola_publica",
                   3:"tp_escola_privada",
                   4:"tp_escola_privada"} #está repetida para ser tratada como tipo 3

  train_copy, test_copy = replace_dummy_remove_column('TP_ESCOLA',dic_TP_ESCOLA, train_copy, test_copy)
  lista_descarte.append('TP_ESCOLA')
  '''
  TP_ENSINO: Tipo de instituição que concluiu ou concluirá o Ensino Médio 
  > 
  '''
  train_copy['TP_ENSINO'] = train_copy['TP_ENSINO'].fillna(1)
  test_copy['TP_ENSINO'] = test_copy['TP_ENSINO'].fillna(1)
  dic_TP_ENSINO = {1:"tp_ensino_n_informado",
                   2:"tp_ensino_regular",
                   3:"tp_ensino_privado",
                   4:"tp_ensino_exterior"}

  train_copy, test_copy = replace_dummy_remove_column('TP_ENSINO',dic_TP_ENSINO, train_copy, test_copy)
  lista_descarte.append('TP_ENSINO')

  '''
  IN_TREINEIRO: Indica se o inscrito fez a prova com intuito de apenas treinar seus conhecimentos
  > 
  '''
  dic_IN_TREINEIRO = {0:"in_treineiro_nao",
                      1:"in_treineiro_sim"}

  train_copy, test_copy = replace_dummy_remove_column('IN_TREINEIRO',dic_IN_TREINEIRO, train_copy, test_copy)
  lista_descarte.append('IN_TREINEIRO')

  '''
  TP_DEPENDENCIA_ADM_ESC: Dependência administrativa (Escola)
  >
  '''
  dic_TP_DEPENDENCIA_ADM_ESC = {0:"tp_escola_n_informada",
                                1:"tp_escola_federal",
                                2:"tp_escola_estadual",
                                3:"tp_escola_municipal",
                                4:"tp_escola_privada"}

  train_copy, test_copy = replace_dummy_remove_column('TP_DEPENDENCIA_ADM_ESC',dic_TP_DEPENDENCIA_ADM_ESC, train_copy, test_copy)  
  lista_descarte.append('TP_DEPENDENCIA_ADM_ESC')
  
  '''
  IN_BAIXA_VISAO: Indicador de baixa visão5
  > Não é necessário modificar a coluna. Nulos serão convertidos em 0.
  '''
  dic_IN_BAIXA_VISAO = {0:"in_baix_visao_nao",
                      1:"in_baix_visa_sim"}

  train_copy, test_copy = replace_dummy_remove_column('IN_BAIXA_VISAO',dic_IN_BAIXA_VISAO, train_copy, test_copy)
  lista_descarte.append('IN_BAIXA_VISAO')


  '''
  IN_CEGUEIRA: Indicador de cegueira
  > Não há pessoas cegas nas amostras, colunas serão descartadas
  '''
  lista_descarte.append('IN_CEGUEIRA')
  
  '''
  IN_SURDEZ: Indicador de surdez
  > Não é necessário modificar a coluna. Nulos serão convertidos em 0.
  '''

  dic_IN_SURDEZ = {0:"in_surdez_nao",
                      1:"in_surdez_sim"}

  train_copy, test_copy = replace_dummy_remove_column('IN_SURDEZ',dic_IN_SURDEZ, train_copy, test_copy)
  lista_descarte.append('IN_SURDEZ')

  '''
  IN_DISLEXIA: Indicador de dislexia
  > Não há dislexicos nas amostras de testes (valor = 1) e apenas 1 na amostra de treino, portanto a coluna e a amostra serão descartados
  '''
  train_copy = train_copy.drop(train_copy.index[train_copy['IN_DISLEXIA'] == 1].tolist(),axis = 0)
  
  lista_descarte.append('IN_DISLEXIA')
  
  '''
  IN_DISCALCULIA: Indicador de discalculia
  > Só há uma amostra com discauculia na amostra de treino e nenhuma na amostra de testes, portanto a coluna e a
  única amostra serão descartadas
  '''
  train_copy = train_copy.drop(train_copy.index[train_copy['IN_DISCALCULIA'] == 1].tolist(),axis = 0)
  
  lista_descarte.append('IN_DISCALCULIA')

  '''
  IN_SABATISTA: Indicador de solicitação de prova aplicada em horário diferenciado para os guardadores dos sábados5
  > 
  '''
  dic_IN_SABATISTA = {0:"in_sabatista_nao",
                      1:"in_sabatista_sim"}

  train_copy, test_copy = replace_dummy_remove_column('IN_SABATISTA',dic_IN_SABATISTA, train_copy, test_copy)
  lista_descarte.append('IN_SABATISTA')
  '''
  IN_GESTANTE: Indicador de gestante
  > 
  '''
  dic_IN_GESTANTE = {0:"in_gestante_nao",
                     1:"in_gestante_sim"}

  train_copy, test_copy = replace_dummy_remove_column('IN_GESTANTE',dic_IN_GESTANTE, train_copy, test_copy)
  lista_descarte.append('IN_GESTANTE')
  '''
  IN_IDOSO: Indicador de inscrito idoso
  > Coluna será emovida por não haver observações com idosos nas duas bases
  '''
  lista_descarte.append('IN_IDOSO')

  '''
  As colunas :  
  TP_PRESENCA_CN: Presença na prova objetiva de Ciências da Natureza;
  TP_PRESENCA_CH: Presença na prova objetiva de Ciências Humanas.

  São iguais e serão unidas representas pela nova coluna TP_PRESENCA_DIA_1
  > 
  '''
  train_copy['TP_PRESENCA_DIA_1'] = train_copy['TP_PRESENCA_CN'] 
  test_copy['TP_PRESENCA_DIA_1'] = test_copy['TP_PRESENCA_CN'] 

  dic_TP_PRESENCA_DIA_1 = {0:"tp_dia_1_faltou",
                        1:"tp_dia_1_presente",
                        2:"tp_dia_1_eliminado"}

  train_copy, test_copy = replace_dummy_remove_column('TP_PRESENCA_DIA_1',dic_TP_PRESENCA_DIA_1, train_copy, test_copy)  
  lista_descarte.append('TP_PRESENCA_DIA_1')
  lista_descarte.append('TP_PRESENCA_CN')
  lista_descarte.append('TP_PRESENCA_CH')
  '''
  TP_PRESENCA_LC: Presença na prova objetiva de Linguagens e Códigos
  > será eliminada, pois só serão consideradas observações com nota maior que 0
  '''
  lista_descarte.append('TP_PRESENCA_LC')
  
  '''
  CO_PROVA_CN: Código do tipo de prova de Ciências da Natureza
  > não é necessário alterar a coluna, serão construídas 9 variáveis dummy
  '''
  
  def dicionario_codigo_prova(codigo):
    append_str = codigo+'_'

    string = np.fromstring(np.arange(train[codigo].unique().shape[0]).tostring(),dtype=int)

    # Append suffix / prefix to strings in list 
    pre_res = [append_str.lower() + str(sub) for sub in string] 

    return {k: v for k, v in zip(train[codigo].unique(),pre_res)}
    
  #train_copy, test_copy = replace_dummy_remove_column('CO_PROVA_CN',dicionario_codigo_prova('CO_PROVA_CN'), train_copy, test_copy) 
  
  lista_descarte.append('CO_PROVA_CN')
  
  '''
  CO_PROVA_CH: Código do tipo de prova de Ciências Humanas
  > não é necessário alterar a coluna, serão construídas 9 variáveis dummy
  '''
  #train_copy, test_copy = replace_dummy_remove_column('CO_PROVA_CH',dicionario_codigo_prova('CO_PROVA_CH'), train_copy, test_copy)
  lista_descarte.append('CO_PROVA_CH')
  
  '''
  CO_PROVA_LC: Código do tipo de prova de Linguagens e Códigos
  > não é necessário alterar a coluna, serão construídas 9 variáveis dummy
  '''
  #train_copy, test_copy = replace_dummy_remove_column('CO_PROVA_LC',dicionario_codigo_prova('CO_PROVA_LC'), train_copy, test_copy)
  lista_descarte.append('CO_PROVA_LC')
  
  '''
  CO_PROVA_MT: Código do tipo de prova de Matemática
  > não é necessário alterar a coluna, serão construídas 9 variáveis dummy
  '''
  #train_copy, test_copy = replace_dummy_remove_column('CO_PROVA_MT',dicionario_codigo_prova('CO_PROVA_MT'), train_copy, test_copy)
  lista_descarte.append('CO_PROVA_MT')
  
  '''
  NU_NOTA_CN: Nota da prova de Ciências da Natureza
  '''
  train_copy['NU_NOTA_CN'] = train_copy['NU_NOTA_CN'].fillna(0)
  test_copy['NU_NOTA_CN'] = test_copy['NU_NOTA_CN'].fillna(0)

  '''
  NU_NOTA_CH: Nota da prova de Ciências Humanas
  '''
  train_copy['NU_NOTA_CH'] = train_copy['NU_NOTA_CH'].fillna(0)#.astype(float)
  test_copy['NU_NOTA_CH'] = test_copy['NU_NOTA_CH'].fillna(0)#.astype(float)

  '''
  NU_NOTA_LC: Nota da prova de Linguagens e Códigos
  '''
  train_copy['NU_NOTA_LC'] = train_copy['NU_NOTA_LC'].fillna(0)
  test_copy['NU_NOTA_LC'] = test_copy['NU_NOTA_LC'].fillna(0)
  '''
  NU_NOTA_MT: Nota da prova de Matemática
  '''
  train_copy['NU_NOTA_MT'] = train_copy['NU_NOTA_MT'].fillna(0)
  
  '''
  TP_LINGUA: Língua Estrangeira 
  > Não é necessário modificar a coluna. Nulos serão convertidos em 0.
  '''
  dic_TP_LINGUA = {0:"tp_lingua_ing",
                   1:"tp_lingua_esp"}

  train_copy, test_copy = replace_dummy_remove_column('TP_LINGUA',dic_TP_LINGUA, train_copy, test_copy)
  lista_descarte.append('TP_LINGUA')
  '''
  TP_STATUS_REDACAO: Situação da redação do participante
  > os valores nulos(NA) serão considerados anulados
  '''
  train_copy['TP_STATUS_REDACAO'] = train_copy['TP_STATUS_REDACAO'].fillna(2)
  test_copy['TP_STATUS_REDACAO'] = test_copy['TP_STATUS_REDACAO'].fillna(2)

  dic_TP_STATUS_REDACAO = {0:"tp_red_anulada",
                           1:"tp_red_sem_problemas",
                           2:"tp_red_anulaad",
                           3:"tp_red_copia_txt_motivador",
                           4:"tp_red_em_branco",
                           5:"tp_red_fere_dir_hum",
                           6:"tp_red_fuga",
                           7:"tp_red_n_atend_tipo",
                           8:"tp_red_txt_insu",
                           9:"tp_red_parte_desconec"}

  train_copy, test_copy = replace_dummy_remove_column('TP_STATUS_REDACAO', dic_TP_STATUS_REDACAO, train_copy, test_copy)
  lista_descarte.append('TP_STATUS_REDACAO')
  '''
  NU_NOTA_COMP1: Nota da competência 1 - Demonstrar domínio da modalidade escrita formal da Língua Portuguesa.
  > variável será descartada
  '''
  lista_descarte.append('NU_NOTA_COMP1')
  '''
  NU_NOTA_COMP2: Nota da competência 2 - Compreender a proposta de redação e aplicar conceitos das várias áreas de conhecimento para desenvolver o tema, dentro dos limites estruturais do texto dissertativo-argumentativo em prosa.
  > variável será descartada 
  '''
  lista_descarte.append('NU_NOTA_COMP2')

  '''
  NU_NOTA_COMP3: Nota da competência 3 - Selecionar, relacionar, organizar e interpretar informações, fatos, opiniões e argumentos em defesa de um ponto de vista.
  > variável será descartada
  '''
  lista_descarte.append('NU_NOTA_COMP3')
  '''
  NU_NOTA_COMP4: Nota da competência 3 - Selecionar, relacionar, organizar e interpretar informações, fatos, opiniões e argumentos em defesa de um ponto de vista.
  > variável será descartada
  '''
  lista_descarte.append('NU_NOTA_COMP4')
  '''
  NU_NOTA_COMP5: Nota da competência 5 - Elaborar proposta de intervenção para o problema abordado, respeitando os direitos humanos.
  > variável será descartada
  '''
  lista_descarte.append('NU_NOTA_COMP5')
  '''
  NU_NOTA_REDACAO: Nota da prova de redação
  >
  '''
  train_copy['NU_NOTA_REDACAO'] = train_copy['NU_NOTA_REDACAO'].fillna(0)
  test_copy['NU_NOTA_REDACAO'] = test_copy['NU_NOTA_REDACAO'].fillna(0)
  '''
  Q001: Até que série seu pai, ou o homem responsável por você, estudou?
  >
  '''
  dic_q001 = {'A':"Q001_A",
              'B':"Q001_B",
              'C':"Q001_C",
              'D':"Q001_D",
              'E':"Q001_E",
              'F':"Q001_F",
              'G':"Q001_G",
              'H':"Q001_H"}

  train_copy, test_copy = replace_dummy_remove_column('Q001',dic_q001, train_copy, test_copy)
  lista_descarte.append('Q001')
  '''
  Q002: Até que série sua mãe, ou a mulher responsável por você, estudou?
  > 
  '''
  dic_q002 = {'A':"Q002_A",
              'B':"Q002_B",
              'C':"Q002_C",
              'D':"Q002_D",
              'E':"Q002_E",
              'F':"Q002_F",
              'G':"Q002_G",
              'H':"Q002_H"}


  train_copy, test_copy = replace_dummy_remove_column('Q002',dic_q002, train_copy, test_copy)
  lista_descarte.append('Q002')
  '''
  Q006: Qual é a renda mensal de sua família? (Some a sua renda com a dos seus familiares.) 
  > 
  '''
  dic_q006 = {'A':"Q006_A",
              'B':"Q006_B",
              'C':"Q006_C",
              'D':"Q006_D",
              'E':"Q006_E",
              'F':"Q006_F",
              'G':"Q006_G",
              'H':"Q006_H",
              'I':"Q006_I",
              'J':"Q006_J",
              'K':"Q006_K",
              'L':"Q006_L",
              'M':"Q006_M",
              'N':"Q006_N",
              'O':"Q006_O",
              'P':"Q006_P",
              'Q':"Q006_Q"}

  train_copy, test_copy= replace_dummy_remove_column('Q006',dic_q006, train_copy, test_copy) 
  lista_descarte.append('Q006')           
  '''
  Q024:Na sua residência tem computador?
  > 
  '''
  dic_q024 = {'A':"Q024_A",
              'B':"Q024_B",
              'C':"Q024_C",
              'D':"Q024_D",
              'E':"Q024_E"}

  train_copy, test_copy = replace_dummy_remove_column('Q024',dic_q024, train_copy, test_copy)  
  lista_descarte.append('Q024')                        

  '''
  Q025: Na sua residência tem acesso à Internet?
  > 
  '''
  dic_q025 = {'A':"Q025_A",
              'B':"Q025_B"}

  train_copy, test_copy = replace_dummy_remove_column('Q025',dic_q025, train_copy, test_copy)
  lista_descarte.append('Q025')

  '''
  Q026: Você exerce ou já exerceu atividade remunerada?
  > 
  '''
  dic_q026 = {'A':"Q026_A",
              'B':"Q026_B",
              'C':"Q026_C"}
  train_copy, test_copy = replace_dummy_remove_column('Q026',dic_q026, train_copy, test_copy)  
  lista_descarte.append('Q026')

  '''
  Q027: Com que idade você começou a exercer uma atividade remunerada?
  >
  '''
  train_copy['Q027'] = train_copy['Q027'].fillna('N')
  test_copy['Q027'] = test_copy['Q027'].fillna('N')
  dic_q027 = {'A':"Q027_A",
              'B':"Q027_B",
              'C':"Q027_C",
              'D':"Q027_D",
              'E':"Q027_E",
              'F':"Q027_F",
              'G':"Q027_G",
              'H':"Q027_H",
              'I':"Q027_I",
              'J':"Q027_J",
              'K':"Q027_K",
              'L':"Q027_L",
              'M':"Q027_M",
              'N':"Q027_N"}#não está no questionário mas representa os nulos, que não trabalham
  train_copy, test_copy = replace_dummy_remove_column('Q027',dic_q027, train_copy, test_copy)    
  lista_descarte.append('Q027')

  '''
  Q047: Em que tipo de escola você frequentou o Ensino Médio?
  >
  '''
  dic_q047 = {'A':"Q047_A",  
              'B':"Q047_B",
              'C':"Q047_C",
              'D':"Q047_D",
              'E':"Q047_E"}

  train_copy, test_copy = replace_dummy_remove_column('Q047',dic_q047, train_copy, test_copy)
  lista_descarte.append('Q047')    

  #remoção das colunas listadas na base de testes
  test_copy = test_copy.drop(lista_descarte,axis=1)

  return train_copy, test_copy



```
